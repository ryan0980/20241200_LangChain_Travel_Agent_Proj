```python
import pandas as pd

# 文件路径
file_path = r"G:\Code\Projects\GWU\24_FA\AML\Final_proj\hotels.csv"

# 加载数据集（加载全部列）
try:
    data = pd.read_csv(
        file_path,
        encoding="Windows-1252",  # 固定使用 Windows-1252 编码
    )
    print("数据集加载成功！")
except Exception as e:
    print(f"数据集加载失败: {e}")
    raise

# 随机抽取 10,000 个记录
sample_size = 10000
sampled_data = data.sample(n=sample_size, random_state=42)  # 设置随机种子保证可复现

# 保存为新的 CSV 文件
output_path = r"G:\Code\Projects\GWU\24_FA\AML\Final_proj\hotels_sampled.csv"
sampled_data.to_csv(output_path, index=False, encoding="Windows-1252")

print(f"随机抽取的 {sample_size} 条数据已保存到：{output_path}")
```
